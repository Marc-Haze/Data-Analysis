# **0 - Descripción del problema**

Se trata de predecir la variable ordinal ***Damage_grade***, que representa un nivel de daño al edificio que fue afectado por el terremoto.

Hay 3 grados de daño: 1 (Daño Bajo), 2 (Daño Medio), 3 (Daño Alto, casi una completa destrucción)

Por lo tanto nos enfrentamos a un problema de ***CLASIFICACIÓN***.

## ***Características***

El conjunto de datos consiste principalmente en información sobre la estructura de los edificios y su propiedad legal. Cada fila del conjunto de datos representa un edificio específico en la región afectada por el terremoto de Gorkha.

Hay 39 columnas en este conjunto de datos, donde la columna building_id es un identificador único y aleatorio. Las 38 funciones restantes se describen en la siguiente sección. Las variables categóricas se ofuscaron con caracteres ascii aleatorios en minúsculas. La aparición del mismo carácter en columnas distintas no implica el mismo valor original.



***· geo_level_1_id, geo_level_2_id, geo_level_3_id (type: int):*** Región geográfica en la que existe el edificio, desde la más grande (nivel 1) hasta la subregión más específica (nivel 3). Valores posibles: nivel 1: 0-30, nivel 2: 0-1427, nivel 3: 0-12567.

***· count_floors_pre_eq (type: int):*** Número de pisos en el edificio antes del terremoto.

***· age (type: int):*** Antigüedad del edificio en años.

***· height_percentage (type: int):*** Altura normalizada de la huella del edificio.

```python
level1_ids = df_train_combined['geo_level_1_id'].unique()
print(f'Conjunto de Datos con Geolocalización de Nivel 1: {level1_ids}')

level2_ids = df_train_combined['geo_level_2_id'].unique()
print(f'Conjunto de Datos con Geolocalización de Nivel 2: {level2_ids}')

level3_ids = df_train_combined['geo_level_3_id'].unique()
print(f'Conjunto de Datos con Geolocalización de Nivel 3: {level3_ids}')
```

*Como queremos entrenarlo con un conjunto de datos variado, tomaremos en cuenta los valores recogidos según el nivel de daños.*

```python
# Crear una lista con los porcentajes para cada valor de damage_grade
# 9% para la parte del dataframe cuyo damage_grade = 1
# 56% para la parte del dataframe cuyo damage_grade = 2
# 33% para la parte del dataframe cuyo damage_grade = 3
porcentajes = [0.09, 0.56, 0.33]

# Crear un diccionario con los dataframes resultantes
df_dict = {}
for i, dg in enumerate([1, 2, 3]):
    df_dict[dg] = df_train_combined.groupby('damage_grade').get_group(dg).sample(frac=porcentajes[i])

# Unir los dataframes en uno solo
train_data_sample = pd.concat(df_dict.values(), axis=0)
train_data_sample.reset_index(drop=True, inplace=True)
```

Una vez esté juntadoos los dos dataframes, tomaremos el 5% del mismo, que serían un total de 12.430  registros con los que entrenar.

```python
# Agrupar los datos por cada nivel geográfico y seleccionar el 10% de los datos aleatorios en cada grupo
train_data_sample = df_train_combined.groupby(["geo_level_1_id", "geo_level_2_id", "geo_level_3_id"]
                                       ).apply(lambda x: x.sample(frac=0.05, random_state=75))

# Resetamos sus índices para ajustar el nuevo conjunto
train_data_sample.reset_index(drop=True, inplace=True)
```

***Separamos los datos de entrenamiento que juntamos para recortarlos y utilizamos el LazyClassifier para ejecutar la clasificación de modelos.***

```python
from sklearn.model_selection import train_test_split
X = train_data_sample.drop(['damage_grade'],axis='columns')
y = train_data_sample['damage_grade']

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.2,random_state=123)
```

## ***Estudio de las Característica con Dendograma***

Utilizaremos un dendograma para visualizar las correlación entre las características.

```python
# Obtenemos la traspuesta de la matriz de las características
X_T = X.T
X_T
```

***Con la matriz traspuesta, generamos el dendograma para corroborar las relaciones entre las características de forma gráfica.***

```python
import sklearn.neighbors
import numpy as np
dist = sklearn.neighbors.DistanceMetric.get_metric('euclidean') 
D = dist.pairwise(X_T)
avD = np.average(D)
print("Distancia Media\t {:6.2f}".format(avD))

# 2. Constuimos el dendograma	
from scipy import cluster	
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
threshold_2 = 7

# Seleccionamos como distancia intercluster el vecino más alejado (complete)
clusters = cluster.hierarchy.linkage(D, method='complete')
cluster.hierarchy.dendrogram(clusters, color_threshold=threshold_2, labels=X.columns.to_list(), leaf_rotation=90)
plt.show()
```

Aún cuando podríamos seleccionar clústeres como agrupaciones de características con cierta altura, primero vamos a ejecutar los modelos utilizando todo el conjunto de características, para así tener una base con la que comprobar a futuro si los procedimientos que se siguen afectan positivamente al modelo.

### ***~ Entrenamiento del Modelo: SVC***

***Uso de GridSearch para parametrizar el modelo y seleccionar el mejor***

```python
from sklearn.model_selection import TimeSeriesSplit,GridSearchCV
from sklearn.svm import SVC

# Seleccionamos el tipo de modelo
svc = SVC()

# Diciconario con parámetros esenciales (24 posibles combinaciones)
params = {'C': [0.1, 1, 10], 
          'kernel': ['linear', 'rbf'], 
          'gamma': ['scale', 'auto'],
          'class_weight': [None, 'balanced']
         }

# Diccionario con parámetros más completos (Hasta 12.000 posibles combinaciones)
# params = {'C': [0.1, 1, 10], 
#           'kernel': ['linear', 'rbf', 'poly'], 
#           'gamma': ['scale', 'auto'],
#           'degree': [2, 3, 4],
#           'coef0': [0.0, 0.5, 1.0],
#           'shrinking': [True, False],
#           'tol': [0.0001, 0.001, 0.01],
#           'max_iter': [-1, 100, 500],
#           'class_weight': [None, 'balanced'],
#           'decision_function_shape': ['ovr', 'ovo']
#          }

# Como saber el número de combinaciones de parámetros que se van a ejecutar
num_combinations = 1
for param in params:
    num_combinations *= len(params[param])
print(f'El número total de combinaciones de parámetros en GridSearchCV es {num_combinations}')

# Utilizamos Cross Validation en el proceso
tscv = TimeSeriesSplit(n_splits=5)

# Generamos el conjunto de modelos con GridSearch y los entrenamos
grid_search = GridSearchCV(estimator = svc, param_grid= params, cv = tscv)
grid_search.fit(X_train, y_train)

print("Mejores Parámetros:", grid_search.best_params_)
print("Puntuación en los Datos de Test:", grid_search.score(X_test, y_test))
```

***Una vez escogido el mejor conjunto de parámetros para nuestro modelo, lo entrenamos y realizamos las predicciones.***

```python
# Guardamos el modelo con mejor parametrización
best_svc = grid_search.best_estimator_

# Entrenamiento y Predicciones del Modelo
best_svc.fit(X_train, y_train)
y_pred = best_svc.predict(X_test)

print(f"Predicciones: {y_pred.astype(int)}")
```

Evaluamos con las métricas del modelo.
*   ***Precisión (Accuracy)***
*   ***Sensibilidad (Recall)***
*   ***Precisión (Precision)***
*   ***Puntuación F1 (F1-Score)***



```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

***Evaluación gráfica de los Resultados***

```python
from sklearn.metrics import f1_score   
f1 = f1_score(y_test, y_pred, average='weighted')   
      
# x axis for plotting
import numpy as np
xx = np.stack(i for i in range(y_test.shape[0]))
plt.plot(xx, y_test, c='r', label='data')
plt.plot(xx, y_pred, c='g', label='prediction')
plt.axis('tight')
plt.legend()
plt.title("SVC con GridSearchCV (F1-Score = %6.2f')" % (f1))
plt.show()
```

```python
plt.scatter(y_test.to_numpy(int), y_pred, alpha=0.3, marker='x')
plt.plot([0,5], [0, 5], "--k")
plt.axis("tight")
plt.xlabel("True Number of Cases")
plt.ylabel("Predicted Number of Cases")
plt.title("SVC con GridSearchCV (F1-Score = %6.2f)" % (f1))
```

### ***~ Entrenamiento del Modelo KNeighborsClassifier***

***Uso de RandomSearch para parametrizar el modelo y seleccionar el mejor***

```python
from sklearn.model_selection import TimeSeriesSplit,RandomizedSearchCV
from sklearn.neighbors import KNeighborsClassifier
from scipy.stats import randint

# Seleccionamos el tipo de modelo
knn = KNeighborsClassifier()

# Diciconario con parámetros esenciales (32 posibles combinaciones)
params = {'n_neighbors': randint(1,30),
          'weights': ['uniform', 'distance'],
          'p': [1, 2],
          'algorithm': ['ball_tree', 'kd_tree']
         }

# Utilizamos Cross Validation en el proceso
tscv = TimeSeriesSplit(n_splits=5)

# Generamos el conjunto de modelos con GridSearch y los entrenamos
random_search = RandomizedSearchCV(knn, params, n_iter=10, cv=tscv, n_jobs=-1)

random_search.fit(X_train, y_train)

print("Mejores Parámetros:", random_search.best_params_)
print("Precisión en los Datos de Test:", random_search.score(X_test, y_test))
```

***Una vez determinado el mejor conjunto de parámetros para el modelo, lo seleccionamos y reentrenamos para evaluarlo.***

```python
# Guardamos el modelo con mejor parametrización
best_knn = random_search.best_estimator_

# Entrenamiento y Predicciones del Modelo
best_knn.fit(X_train, y_train)
y_pred = best_knn.predict(X_test)
```

Evaluamos sus métricas.

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

### ***~ Entrenamiento del Modelo DecisionTreeClassifier***

***Usamos GridSearch para parametrizar nuestro modelo de árboles.***

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit

# Definir el diccionario con los posibles valores que 
# se probarán para cada parámetro del modelo (375 combinaciones)
params = {'max_depth': [2, 4, 6, 8, 10],
          'min_samples_split': [2, 4, 6, 8, 10],
          'min_samples_leaf': [1, 2, 4, 6, 8],
          'max_features': ['auto', 'sqrt', 'log2']
         }

# Crear un modelo de árboles de decisión para clasificación
rf = RandomForestClassifier()

# Utilizamos Cross Validation en el proceso
tscv = TimeSeriesSplit(n_splits=5)

# Como saber el número de combinaciones de parámetros que se van a ejecutar
num_combinations = 1
for param in params:
    num_combinations *= len(params[param])
print(f'El número total de combinaciones de parámetros en GridSearchCV es {num_combinations}')

# Realizar una búsqueda de cuadrícula para encontrar los mejores parámetros
grid_search = GridSearchCV(rf, params, cv=tscv, n_jobs=-1)

# Entrenamos el conjunto de modelos
grid_search.fit(X_train, y_train)


# Imprimir los mejores parámetros y el mejor resultado
print("Mejores parámetros:", grid_search.best_params_)
print("Mejor resultado:", grid_search.best_score_)
```

***Una vez determinado el mejor conjunto de parámetros, entrenamos nuestro modelo.***

```python
# Guardamos el modelo con mejor parametrización
best_forest = grid_search.best_estimator_

# Entrenamiento y Predicciones del Modelo
best_forest.fit(X_train, y_train)
y_pred = best_forest.predict(X_test)

print(f"Predicciones: {y_pred.astype(int)}")
```

**Evaluamos las métricas de nuestro modelo**

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

**Representación gráfica de los resultados**

```python
from sklearn.metrics import f1_score   
f1 = f1_score(y_test, y_pred, average='weighted')   
      
# x axis for plotting
import numpy as np
xx = np.stack(i for i in range(y_test.shape[0]))
plt.plot(xx, y_test, c='r', label='data')
plt.plot(xx, y_pred, c='g', label='prediction')
plt.axis('tight')
plt.legend()
plt.title("RandomForestClassifier con GridSearchCV (F1-Score = %6.2f')" % (f1))
plt.show()
```
